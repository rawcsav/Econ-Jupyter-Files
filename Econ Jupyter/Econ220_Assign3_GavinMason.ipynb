```python
# Import packages here 
import scipy.stats as stats
from scipy.stats import norm 
from scipy.stats import uniform 
import random
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Practice with Random Number Generation and Histograms <a class = anchor id = anchor1></a>

Play around with this functions, start by looking at the documentation in Python.
* Set the seed at 245
* Generate a vector y of 100 observations from the normal distribution with mean=3 and standard deviation sd=7 
* Generate a vector z of 100 observations from the normal distribution with mean=30 and sd=3
* Generate a vector u of 100 observations from the uniform distribution  
* Generate a vector v of 1000 observations of integers between 0 to 5000. The function random.sample() is adequate for this task. But you might use others if you want.
* Generate a vector v_sample which a sample of 100 observations of without replacement taken from v

```python
np.random.seed(245)
y = norm.rvs(loc=3, scale=7, size=100)
z = norm.rvs(loc=30, scale=3, size=100)
u = uniform.rvs(size=100)
v = random.sample(range(0,5000), 1000)
v_sample = np.random.choice(v, size=100, replace=False)
```

* Make a 2 histograms of y
    * one without any changes from the default options. 
    * Make another histogram of y where you change the number of bins to 20, add a density curve and change the color or histogram to "green" and add a title and labels

```python
data = pd.DataFrame({"y":y, "z":z, "u":u})
sns.displot([data.y])
```

```python
sns.displot([data.y], kde=True, bins=20, palette=["green"]).set(title="Histogram of y")
plt.ylabel("Freq and Density")
```


* Make a histogram of z and y in the same graph add a title and change colors (you can choose) use the  `sns.displot` for this 

```python
sns.displot([data.y, data.z], kde=True, palette=["blue", "green"]).set(title="Histogram of y and z")
plt.ylabel("Freq and Density")
```

* Make a histogram of u and change the number of bins= 20, change colors to your choice add titles and labels 

```python
sns.displot([data.u], bins=20, kde=True, palette=["purple"]).set(title="Histogram of u")
plt.ylabel("Freq and Density")
```

# A few probability exercises

Probability that a normal random variable with mean 10 and variance 49

* (i) lies between 9 and 12
* (ii) is greater than 25
* (iii) is less than 13

```python
p1 = stats.norm.cdf(12, 10, 7) - stats.norm.cdf(9, 10, 7)
p2 = 1 - stats.norm.cdf(25, 10, 7)
p3 = stats.norm.cdf(13, 10, 7)
print("Prob(9<x<12)=", p1)
print("Prob(x>25)=", p2)
print("Prob(x<13)=", p3)
```
